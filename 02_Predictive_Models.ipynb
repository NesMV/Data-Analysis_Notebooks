## Linear Regression
Linear regression is fitting a straight line to a set of observation. For example, let's say that I have a bunch of people that I measured and the two features that I measured of these people are their weight and height:



The equation of a stright line is y = mx + b

## The Ordinary Least Squares technique (OLS)

How does linear regression work?

* The way it works is it tries to minimize the squared error between each point and the line, where the error is just the distance between each point and the line that you have.

* So, we sum up all the squares of those errors, which sounds a lot like when we computed variance, except that instead of relative to the mean, it's relative to the line that we're defining. We can measure that variance of the data points from the line, and by minimizing that variance, we can find the line that fits it the best.

## The co-efficient of determination or r-squared
How do I know good my regression is? That's where r-squared comes in, and r-squared is also known as the coefficient of determination.

## Computing r-squared

1.0 - (sum_of_squared_errors / sum_of_squared_variation_from_mean)

For r-squared. you will get a value that ranges from 0 to a 1.

* 0 means your fit is terrible. 
* 1 means your fit is perfect.
